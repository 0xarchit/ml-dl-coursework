```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
import requests
def load(url):
    filename = os.path.basename(url)
    response = requests.get(url)
    with open(filename, "wb") as f:
        f.write(response.content)
    df = pd.read_csv(filename)
    display(df.head())
    return df
```

```python
url="https://files.0xarchit.is-a.dev/datasets/pca_data.csv"
df = load(url)
```

```python
# split input and output variable
X = df.drop("Label", axis=1)
y = df["Label"]
```

```python
sns.kdeplot(df["Unnamed: 6"], fill=True)
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import explained_variance_score, silhouette_score
```

```python
pca = PCA(n_components=5, random_state=21)
```

```python
x_pca = pca.fit_transform(X) # if only fit then class will be pca and cant be split directly so transform necessary
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
sns.lineplot(df['Label'])
```

```python
type(x_pca)
```

```python
x_pca_df = pd.DataFrame(x_pca, columns=['pc1', 'pc2', 'pc3', 'pc4', 'pc5'])
y_df = pd.DataFrame(y)
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_pca_df, y_df, train_size=0.8, random_state=33)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, precision_score, recall_score
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, Y_train)
```

```python
Y_pred = lr_model.predict(X_test)
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
f1_score(Y_test, Y_pred, average='weighted')
```

```python
precision_score(Y_test, Y_pred, average='weighted')
```

```python
recall_score(Y_test, Y_pred, average='weighted')
```

```python
# kmeans cluster
from sklearn.cluster import KMeans, DBSCAN
```

```python
kmeans = KMeans(n_clusters=2, random_state=3)
```

```python
kmeans.fit(X_train)
```

```python
Y_pred_kmeans = kmeans.predict(X_test)
```

```python
silhouette_score(X_test, Y_pred_kmeans)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
```

```python
dbscan.fit(X_train)
```

```python
Y_pred_dbscan = dbscan.fit_predict(X_test)
```

```python
silhouette_score(X_test, Y_pred_dbscan)
```
